```python
from faker import Faker
import pandas as pd
import random

fake = Faker()

def generate_patient_record(i):
    id = i
    name = fake.name()
    age = random.randint(4, 90)
    if age < 18:
        height = random.randint(100, 150)  # Height between 100-150 cm for age below 18
        weight = random.randint(60, 100)      # Weight less than 50 lb for age below 18
    else:
        height = random.randint(150, 200)  # Height between 150-200 cm for age 18 and above
        weight = random.randint(100, 220)   # in lb
    
    gender = random.choice(['Male', 'Female'])
    bloodGroup = random.choice(['O+', 'O-', 'A+', 'B+',])
    
    if gender == 'Male':
        pregnancy = 'No'
    else:
        pregnancy = random.choice(['Yes', 'No'])
    
    past_medical_conditions = random.choice(["Acne", "Asthama", "Cancer", "Covid19", "Diabetes(type 1)", "Diabetes(type 2)", "Allergies", "Depression", "Hayfever", "Stroke", "Colds & Flu", "Hypertension"])  # Generating a random sentence for past medical conditions
    return {
        'Patient ID': id,
        'Name': name,
        'Age': age,
        'Gender': gender,
        'Height(cm)': height,
        'Weight(lb)': weight,
        'Pregnancy': pregnancy,
        'Blood Group': 
        'Past Medical Conditions': past_medical_conditions
    }

def generate_patient_records(num_records=10):
    records = []
    for i in range(num_records):
        records.append(generate_patient_record(i))
    return records

# Generating 10 patient records
patient_records = generate_patient_records(2000)

df = pd.DataFrame(patient_records)
```
